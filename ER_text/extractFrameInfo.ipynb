```python
IN_PATH = '/content/data/'
OUT_PATH= '/content/frame_data/'
```

```python
def label_extract(in_path, out_path):

  json_list = sorted(glob.glob(in_path+'*/*/*.json'))
  data_dict = {}

  for json_name in tqdm(json_list): # json 순회
      with open(json_name, "r") as f:
          try:
              json_file = json.load(f, encoding="utf-8")
          except:
              print("Response content is not valid JSON")
              continue

      file_num = int(json_name.split('/')[-1].replace('clip_', '').replace('.json', '')) # json 번호
      data_dict = {} # json 번호별 data 저장 dict

      if json_file.get('data'): # json data
          # frame 번호 순으로 정렬
          frame_list = list(json_file['data'].keys())
          frame_list = list(map(int, frame_list))
          frame_list.sort()
          frame_list = list(map(str, frame_list))


          for frame in frame_list:
              data_dict[int(frame)] = {}
              if json_file['data'][frame].get('1'):
                  if json_file['data'][frame]['1'].get('emotion'): # 감정 포함된 경우 해당 key 정보 저장
                      data_dict[int(frame)][1] = json_file['data'][frame]['1']
              if json_file['data'][frame].get('2'):
                  if json_file['data'][frame]['2'].get('emotion'): # 감정 포함된 경우 해당 key 정보 저장
                      data_dict[int(frame)][2] = json_file['data'][frame]['2']

              if not data_dict[int(frame)]:
                  del(data_dict[int(frame)])

          with open(out_path + f'clip_{file_num}.json', 'w', encoding='utf-8') as frame_data: # 해당 json 번호 frame data 저장

              json.dump(data_dict, frame_data, indent="\t", ensure_ascii=False)

```

```python
label_extract(IN_PATH, OUT_PATH)
```
